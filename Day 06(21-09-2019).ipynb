### Data Cleaning

- isnull() will notify the any missing values in the dataframe

```python
# Entire data set
df.isnull().sum()
```

```python
# Only column missing values
df['Season'].isnull().sum()
```

- **dropna()** This function allows you to drop all or some rows that have missing values

```python
len(df)
```

### Visualizing Data
- An intersting way to display Dataframe
- With matplotlib

```python
import matplotlib.pyplot as plt
a1 = [1,2,3,4]
a2 = [5,6,7,8]
plt.figure(figsize=(5,5))
plt.plot(a1,a2,color='green')
plt.xlabel('x-Axis')
plt.ylabel('y-Axis')
plt.show()
```

```python
divisions = [1,2,3,4]
marks = [79,82,75,63]
avg_marks = [65,72,78,89]
width = 0.70
plt.bar(divisions,marks,color="blue",width=0.3)
plt.show()
```

```python
plt.barh(divisions,marks,color='blue')
plt.show()
```

```python
lang = ['Python','Java','C','C++']
rating = [75,70,55,65]
plt.pie(rating,labels = lang,shadow=True,startangle=90,autopct='%1.1f%%')
plt.legend(loc='best')
plt.show()
```
